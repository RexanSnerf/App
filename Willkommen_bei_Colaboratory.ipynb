<h1>Was ist Colaboratory?</h1>

Mit Colaboratory oder kurz "Colab" können Sie Python-Code in Ihrem Browser schreiben und ausführen. Sie können Folgendes tun: 
- Keine Konfiguration erforderlich
- Kostenlosen Zugriff auf GPUs
- Einfache Freigabe

Egal, ob Sie <strong>Student</strong>, <strong>Data Scientist</strong> oder <strong>AI-Forscher</strong> sind – Colab erleichtert Ihnen die Arbeit. Im Video <a href="https://www.youtube.com/watch?v=inN8seMm7UI">Einführung in Colab</a> erhalten Sie weitere Informationen, Sie können aber auch gleich hier loslegen.

## <strong>Erste Schritte</strong>

Das Dokument, das Sie lesen, ist keine statische Webseite, sondern eine interaktive Umgebung, die als <strong>Colab-Notebook</strong> bezeichnet wird und in der Sie Code schreiben und ausführen können.

Hier ist beispielsweise eine <strong>Codezelle</strong> mit einem kurzen Python-Skript, das einen Wert berechnet, ihn in einer Variablen speichert und das Ergebnis ausgibt:

```
seconds_in_a_day = 24 * 60 * 60
seconds_in_a_day
```

Wählen Sie zum Ausführen des Codes in der Zelle oben die Zelle mit einem Klick aus und drücken Sie dann entweder die Schaltfläche zum Abspielen links neben dem Code oder verwenden Sie die Tastenkombination "Befehlstaste/Strg + Enter". Variablen, die Sie in einer Zelle definieren, können später in anderen Zellen verwendet werden:

```
seconds_in_a_week = 7 * seconds_in_a_day
seconds_in_a_week
```

## Data Science

Mit Colab können Sie das volle Potenzial der beliebten Python-Bibliotheken nutzen, um Daten zu analysieren und visualisieren. Die Codezelle unten verwendet <strong>NumPy</strong>, um zufällige Daten zu erstellen, und <strong>Matplotlib</strong>, um sie zu visualisieren. Klicken Sie zum Bearbeiten des Codes einfach auf die Zelle und beginnen Sie mit der Bearbeitung.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

## Beispiele für maschinelles Lernen

Wenn Sie Interesse an End-to-End-Beispielen für interaktive Analysen mit maschinellem Lernen haben, wie sie mit Colaboratory möglich werden, sehen Sie sich die folgenden Anleitungen an, bei denen Modelle von <a href="https://tfhub.dev">TensorFlow Hub</a> zum Einsatz kommen.

Hier einige Beispiele:

- <a href="https://tensorflow.org/hub/tutorials/tf2_image_retraining">Erneutes Trainieren eines Bildklassifikators</a>: Erstellen Sie auf Grundlage eines zuvor trainierten Bildklassifikators ein Keras-Modell, mit dem Blumen auseinandergehalten werden können.
- <a href="https://tensorflow.org/hub/tutorials/tf2_text_classification">Textklassifizierung</a>: Klassifizieren Sie Filmrezensionen in der IMDB als <em>positiv</em> oder <em>negativ</em>.
- <a href="https://tensorflow.org/hub/tutorials/tf2_arbitrary_image_stylization">Stilübertragung</a>: Übertragen Sie den Stil eines Bildes, mithilfe von Deep Learning, auf ein anderes Bild.
- <a href="https://tensorflow.org/hub/tutorials/retrieval_with_tf_hub_universal_encoder_qa">Fragen und Antworten zum mehrsprachigen universellen Satz-Encoder</a>: Verwenden Sie zum Beantworten von Fragen aus dem SQuAD-Dataset ein Modell für maschinelles Lernen.
- <a href="https://tensorflow.org/hub/tutorials/tweening_conv3d">Videointerpolation</a>: Machen Sie eine Vorhersage darüber, was zwischen dem ersten und dem letzten Bild eines Videos geschah.


```
import tensorflow as tf
from tensorflow import keras

model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer="sgd", loss="mean_squared_error")

xs=[1, 2, 3]
ys=[2, 4, 6]

model.fit(xs, ys, epochs=1000)

print(model.predict([7]))
```
